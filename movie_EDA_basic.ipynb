## 요약본

````python
def resumtable(data):
    summary = pd.DataFrame(data.dtypes).reset_index().rename(columns={"index":"피처", 0:"dtypes"})
    summary["결측치 갯수"] = data.isnull().sum().values
    summary["고윳값 갯수"] = data.nunique().values
    summary["첫번째 값"] = data.iloc[0].values
    summary["두번째 값"] = data.iloc[1].values
    summary["세번째 값"] = data.iloc[2].values
    
    return summary
````

## 영화제목 중복 정리

````python
mask = df["영화제목"].value_counts()>1 # 중복되는 영화 확인
duplicated_movie = df["영화제목"].value_counts()[mask].index # 중복되는 영화제목 추출
cols = ["매출금액", "관람객", "서울매출금액", "서울관람객수"] # 중복영화에서 합해야할 피쳐
len(duplicated_movie)
````

````python
df.loc[df["영화제목"] == "코다"]
````

````python
for movie in duplicated_movie: # 중복된 영화 [매출금액, 관람객, 서울매출금액, 서울관람객수] 합치기
    for col in cols:
        if df.loc[df["영화제목"] == movie, col].isna().all():
            value = None # 결측치만 있는 피쳐는 그대로 결측치 유지
        else:
            value = df.loc[df["영화제목"] == movie, col].sum(skipna=True) # 결측치 아닌 것들만 더하기
        
        df.loc[df["영화제목"] == movie, col] = value # 결과 업데이트
        
df.loc[df["영화제목"] == "코다"]
````

````python
df.loc[df["개봉날짜"].str.contains("-", na=False), "개봉날짜"] = (
    df.loc[df["개봉날짜"].str.contains("-", na=False), "개봉날짜"]
    .str.replace("-", "")
) # 개봉날짜에서 "-"기호 없애기

df[df["개봉날짜"].str.contains("-", na=False)]["개봉날짜"].unique()
````
